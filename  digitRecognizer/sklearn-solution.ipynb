# Load data

```python
train_data = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test_data = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
```

# Preparing data by splitting it to y(labels) and X(features)

```python
train_data.describe()
```

```python
y=train_data['label']
X=train_data.copy()
del X['label']
print(X)
np.shape(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
from sklearn.preprocessing import normalize
X_train_norm=normalize(X_train)
X_test_norm=normalize(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

# Init MLP

```python
clf=MLPClassifier(solver='adam',hidden_layer_sizes=350,alpha=1e-04)
```

```python
clf.fit(X_train_norm,y_train)
```

```python
from sklearn.metrics import accuracy_score

preds=clf.predict(X_test_norm)
print(accuracy_score(y_test,preds))
```

```python
X_pred=test_data
```

```python
from sklearn.preprocessing import normalize

result = clf.predict(normalize(X_pred))

df=pd.DataFrame(result)
df.index+=1
print(result.shape)
filename = 'DigitalPredictions2.csv'
df.to_csv(filename,index=True,header=["Label"],index_label=["ImageId"])
print('Saved file: ' + filename)
```
